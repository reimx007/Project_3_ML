```python
import os
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file = "../data/state_min_wages_2014-20.csv"
```

```python
wage_data_df = pd.read_csv(data_file)
wage_data_df
```

```python
wage_data_df.to_csv("../data/state_min_wages.csv")
```

```python
wage_data_df.dtypes
```

```python
wage_data_df["2020"] = wage_data_df["2020"].astype(float)
wage_data_df.dtypes
```

```python
nat_avg_2014 = wage_data_df['2014'].mean()
nat_avg_2014 = round(nat_avg_2014, 2)
nat_avg_2014
```

```python
nat_avg_2015 = wage_data_df['2015'].mean()
nat_avg_2015 = round(nat_avg_2015, 2)
nat_avg_2015
```

```python
nat_avg_2016 = wage_data_df['2016'].mean()
nat_avg_2016 = round(nat_avg_2016, 2)
nat_avg_2016
```

```python
nat_avg_2017 = wage_data_df['2017'].mean()
nat_avg_2017 = round(nat_avg_2017, 2)
nat_avg_2017
```

```python
nat_avg_2018 = wage_data_df['2018'].mean()
nat_avg_2018 = round(nat_avg_2018, 2)
nat_avg_2018
```

```python
nat_avg_2019 = wage_data_df['2019'].mean()
nat_avg_2019 = round(nat_avg_2019, 2)
nat_avg_2019
```

```python
nat_avg_2020 = wage_data_df['2020'].mean()
nat_avg_2020 = round(nat_avg_2020, 2)
nat_avg_2020
```

```python
nat_avg_6yr = [nat_avg_2014, nat_avg_2015, nat_avg_2016, nat_avg_2017, nat_avg_2018, nat_avg_2019, nat_avg_2020]
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020]
x = years
y = nat_avg_6yr
plt.plot(x, y)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Avg Min Wage ($)", fontsize=14)
plt.title("National Average of State Min. Wages over 6 Years")
plt.show()
```

```python
nat_avg_df = pd.DataFrame(nat_avg_6yr, years)
nat_avg_df
nat_avg_df = nat_avg_df.rename_axis("Year", index='columns')
nat_avg_df = nat_avg_df.rename(columns={0: "Avg Min Wage"})
nat_avg_df
```

```python
n = 6
LV = 9.02
FV = 7.59
CompoundAnnualGrowthRate = (((LV/FV) ** (1/n)) - 1) * 100
print(round(CompoundAnnualGrowthRate,1))
```

Minimum wage increased by a rate of 2.9% over the six-year period from 2014-2020.
